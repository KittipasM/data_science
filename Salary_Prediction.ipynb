#Find Min/Max salary of each Jobs

```python
df
```

```python
dfm = df[['Job Title', 'Salary Estimate']]
dfm
```

```python
dfm.info()
```

```python
import re
```

```python
dfm['List of salary'] = dfm['Salary Estimate'].apply(lambda x: re.findall(r'\d+', x))
```

```python
dfm['List of salary'] = dfm['List of salary'].apply(lambda x: x*2 if len(x) == 1 else x)
```

```python
dfm['min_salary_USD'] = (dfm['List of salary'].apply(lambda x : x[0]).astype(int))\
        .apply(lambda x : x*1000 if x != 1 else 'Not Defined')
```

```python
dfm['max_salary_USD'] = (dfm['List of salary'].apply(lambda x : x[1]).astype(int))\
        .apply(lambda x : x*1000 if x != 1 else 'Not Defined')
```

```python
dfm = dfm.loc[:,dfm.columns.str.match(r'[^List+]')]
dfm
```

#Top 10 Highest salary

```python
dfm['max_salary_USD'] = dfm['max_salary_USD'].apply(lambda x : 0 if x=='Not Defined' else x).astype(int)
```

```python
dfm['min_salary_USD'] = dfm['min_salary_USD'].apply(lambda x : 0 if x=='Not Defined' else x).astype(int)
```

```python
dfm.apply(lambda x : (x['min_salary_USD']+x['max_salary_USD'])/2, axis=1)
```

```python
dfm['average_salary_USD'] = dfm.apply(lambda x : (x['min_salary_USD']+x['max_salary_USD'])/2, axis=1)
```

```python
dfm.sort_values(by='average_salary_USD', ascending=False).head(10)
```
